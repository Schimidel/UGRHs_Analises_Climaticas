```python
import os
import re
import xarray as xr
import pandas as pd

# Listagem dos arquivos NetCDF
arquivos = [arq for arq in os.listdir(basedir) if arq.endswith('.nc')]

# Regex para parsear o nome do arquivo
padrao = r'^(.*?)_anom_(precip|temp)_(rcp\d{2})\.nc$'

# Lista para armazenar DataFrames
dfs = []

print(f"🔍 Processando {len(arquivos)} arquivos...")

for i, arq in enumerate(arquivos, start=1):
    match = re.match(padrao, arq)
    if not match:
        print(f'⚠️ Arquivo não corresponde ao padrão: {arq}')
        continue

    modelo, variavel, scenario = match.groups()

    print(f'[{i}/{len(arquivos)}] Processando -> Modelo: {modelo} | Cenário: {scenario} | Variável: {variavel}')

    # Abrindo o NetCDF
    ds = xr.open_dataset(os.path.join(basedir, arq))

    # Detectando a variável correta
    if 'pr' in ds.data_vars:
        dados = ds['pr']
    elif 'tas' in ds.data_vars:
        dados = ds['tas']
    else:
        raise ValueError(f'❌ Variável esperada não encontrada em {arq}')

    # Média espacial, se tiver lat/lon
    if set(['lat', 'lon']).issubset(dados.dims):
        dados = dados.mean(dim=['lat', 'lon'])

    # Verificando nome do eixo temporal (pode ser 'time' ou 'year')
    time_dim = 'time' if 'time' in dados.dims else 'year'

    # Extrair anos a partir de datas se for time
    if time_dim == 'time':
        anos = pd.to_datetime(dados[time_dim].values).year
    else:
        anos = dados[time_dim].values

    # Criar DataFrame
    df = pd.DataFrame({
        'year': anos,
        'value': dados.values
    })

    # Adicionar metadados
    df['model'] = modelo
    df['scenario'] = scenario
    df['variable'] = 'precipitation' if variavel == 'precip' else 'temperature'

    dfs.append(df)

print("✅ Todos os arquivos foram processados!")

# Concatenar todos
df_total = pd.concat(dfs)

# Pivotar para ter precipitação e temperatura como colunas
df_final = df_total.pivot_table(
    index=['year', 'model', 'scenario'],
    columns='variable',
    values='value',
    aggfunc='mean'
).reset_index()

# Corrigir nomes
df_final.columns.name = None

print("DataFrame final criado com sucesso!")

```

```python
df_final
```

```python
# Cálculo da diferença ano a ano (diff) por modelo e cenário
df_diff = df_final.copy()

df_diff['precipitation_diff'] = df_diff.groupby(['model', 'scenario'])['precipitation'].diff()
df_diff['temperature_diff'] = df_diff.groupby(['model', 'scenario'])['temperature'].diff()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 🔧 Preparação dos dados
df_pca = df_diff.dropna(subset=['precipitation_diff', 'temperature_diff']).copy()

X = df_pca[['precipitation_diff', 'temperature_diff']]

# Padronizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

# Construir DataFrame com resultados da PCA
df_pca_result = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
df_pca_result = pd.concat([df_pca_result, df_pca[['model', 'scenario']].reset_index(drop=True)], axis=1)

# Variância explicada
explained_var = pca.explained_variance_ratio_

# Loadings (vetores dos eixos)
loadings = pca.components_.T  # shape: (n_features, n_components)
features = ['precipitation_diff', 'temperature_diff']
```

```python
# Plot
plt.figure(figsize=(8, 6))

# Plot dos pontos
sns.scatterplot(
    data=df_pca_result,
    x='PC1', y='PC2',
    hue='scenario',
    style='model',
    s=80
)

# Plot dos vetores dos eixos
for i, feature in enumerate(features):
    plt.arrow(0, 0,
              loadings[i, 0] * 2,  # Escala dos vetores
              loadings[i, 1] * 2,
              color='black',
              alpha=0.8,
              head_width=0.05)
    plt.text(loadings[i, 0] * 2.2,
             loadings[i, 1] * 2.2,
             feature,
             color='black',
             ha='center', va='center', fontsize=10)

# Labels e estética
plt.title(f'PCA das Diferenças (PC1 {explained_var[0]*100:.1f}% | PC2 {explained_var[1]*100:.1f}%)')
plt.xlabel(f'PC1 ({explained_var[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({explained_var[1]*100:.1f}%)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Interpretação do Gráfico de PCA

O que o gráfico sugere é o seguinte:

- Os vetores de `precipitation_diff` e `temperature_diff` estão aproximadamente ortogonais (em torno de 90 graus).
- Isso indica que há **pouca ou nenhuma correlação linear** entre as diferenças ano a ano de precipitação e temperatura nos dados.
- A **PC1 (76%)** está fortemente associada a uma única variável — olhando para a orientação dos vetores, possivelmente a **precipitação**.
- A **PC2 (24%)** representa quase todo o restante da variabilidade, provavelmente associada à **temperatura** ou a **ruído estatístico**.

### Teste de Hipótese: Co-variação entre Diferenças de Precipitação e Temperatura

**Objetivo:** Avaliar se as diferenças interanuais de precipitação (`precipitation_diff`) co-variam com as diferenças interanuais de temperatura (`temperature_diff`).

### Formulação do Teste

- **Hipótese Nula (H₀):**  
Não existe co-variação significativa entre as diferenças anuais de precipitação e temperatura.  
As variáveis são estatisticamente independentes.

- **Hipótese Alternativa (H₁):**  
Existe co-variação significativa entre as diferenças anuais de precipitação e temperatura.  
As variáveis não são independentes.

**Métodos Sugeridos**

1. **Correlação de Pearson (linear)**  
   Testa se há correlação linear entre `precipitation_diff` e `temperature_diff`.

2. **Correlação de Spearman (monotônica, não linear)**  
   Testa se há uma relação monotônica (não necessariamente linear) entre as variáveis.

3. **Regressão Linear Simples**  
   Avalia se as variações de temperatura explicam as variações de precipitação.

4. **Análise de Correlação Cruzada (CCF)**  
   Testa se há correlação com defasagem temporal (lag), ou seja, se a precipitação responde às variações de temperatura com atraso.

5. **Análise por Subgrupos (Modelo e Cenário)**  
    Executa os testes acima separadamente para cada modelo e cenário, identificando se a co-variação ocorre apenas em alguns contextos.

**Interpretação dos Resultados**

- **Se p-valor < 0.05:**  
Rejeitamos H₀. Existe evidência estatística de que as diferenças de precipitação e temperatura co-variam.

- **Se p-valor ≥ 0.05:**  
Não rejeitamos H₀. Não há evidência suficiente de co-variação significativa entre as variáveis.



```python
from scipy.stats import pearsonr
import pandas as pd

# Lista para armazenar os resultados
resultados = []

# Loop por modelo e cenário
for (modelo, scenario), grupo in df_diff.groupby(['model', 'scenario']):
    # Remover NaN
    dados = grupo.dropna(subset=['precipitation_diff', 'temperature_diff'])

    if len(dados) > 1:  # Garante que há dados suficientes
        corr, p = pearsonr(dados['precipitation_diff'], dados['temperature_diff'])
    else:
        corr, p = (None, None)  # Caso tenha poucos dados

    resultados.append({
        'Model': modelo,
        'Scenario': scenario,
        'Correlation': corr,
        'p-value': p
    })

# Converter para DataFrame
df_corr = pd.DataFrame(resultados)

# Opcional: arredondar para exibir
df_corr['Correlation'] = df_corr['Correlation'].round(3)
df_corr['p-value'] = df_corr['p-value'].round(4)

df_corr

```

### Interpretação dos Resultados

- Existe uma correlação negativa estatisticamente significativa entre as diferenças anuais de temperatura e precipitação em todos os modelos e cenários.
- A correlação varia de -0.37 a -0.735, sendo mais forte no modelo HadGEM2_ES, especialmente no cenário rcp85.
- Isso indica que anos com aumento de temperatura tendem a apresentar redução na precipitação.
- O padrão é consistente entre modelos e mais intenso sob maior forçamento radiativo (rcp85).


```python
from scipy.stats import spearmanr
import pandas as pd

# Lista para armazenar os resultados
resultados_spearman = []

# Loop por modelo e cenário
for (modelo, scenario), grupo in df_diff.groupby(['model', 'scenario']):
    # Remover NaN
    dados = grupo.dropna(subset=['precipitation_diff', 'temperature_diff'])

    if len(dados) > 1:  # Garante que há dados suficientes
        corr, p = spearmanr(dados['precipitation_diff'], dados['temperature_diff'])
    else:
        corr, p = (None, None)

    resultados_spearman.append({
        'Model': modelo,
        'Scenario': scenario,
        'Correlation': corr,
        'p-value': p
    })

# Converter para DataFrame
df_spearman = pd.DataFrame(resultados_spearman)

# Arredondar os resultados
df_spearman['Correlation'] = df_spearman['Correlation'].round(3)
df_spearman['p-value'] = df_spearman['p-value'].round(4)

df_spearman
```

### Correlação de Spearman - Interpretação dos Resultados

- Todos os modelos e cenários apresentam correlação negativa significativa entre as diferenças anuais de precipitação e temperatura.
- As correlações variam de -0.396 a -0.679, sendo mais fortes no modelo HadGEM2_ES, especialmente no cenário rcp85.
- Isso indica que, de forma monotônica, anos com aumento de temperatura estão associados à redução na precipitação.

### Diferença entre Spearman e Pearson

- **Pearson** mede correlação linear e é sensível a relações diretamente proporcionais.
- **Spearman** mede correlação monotônica, baseada na ordenação dos dados, e é mais robusta contra outliers e relações não lineares.
- A semelhança dos resultados entre Spearman e Pearson sugere que a relação negativa entre variações de temperatura e precipitação é consistente tanto de forma linear quanto monotônica.


```python
import statsmodels.api as sm
import pandas as pd

# Lista para armazenar os resultados
resultados_regressao = []

# Loop por modelo e cenário
for (modelo, scenario), grupo in df_diff.groupby(['model', 'scenario']):
    dados = grupo.dropna(subset=['precipitation_diff', 'temperature_diff'])

    if len(dados) > 1:
        X = sm.add_constant(dados['temperature_diff'])
        y = dados['precipitation_diff']

        modelo_ols = sm.OLS(y, X).fit()

        resultados_regressao.append({
            'Model': modelo,
            'Scenario': scenario,
            'Coef_Temp': modelo_ols.params['temperature_diff'],
            'Intercept': modelo_ols.params['const'],
            'R-squared': modelo_ols.rsquared,
            'p-value': modelo_ols.pvalues['temperature_diff']
        })
    else:
        resultados_regressao.append({
            'Model': modelo,
            'Scenario': scenario,
            'Coef_Temp': None,
            'Intercept': None,
            'R-squared': None,
            'p-value': None
        })

# Converter para DataFrame
df_regressao = pd.DataFrame(resultados_regressao)

# Arredondar os resultados
df_regressao = df_regressao.round({'Coef_Temp': 3, 'Intercept': 3, 'R-squared': 3, 'p-value': 4})

df_regressao
```

### Regressão Linear Simples - Interpretação dos Resultados

- Todos os modelos e cenários apresentam coeficiente negativo para temperatura, indicando que aumentos nas diferenças de temperatura estão associados a reduções nas diferenças de precipitação.
- O coeficiente varia de -90.404 a -188.500, sendo mais intenso no modelo ACCESS1_3 (rcp85) e HadGEM2_ES (rcp85).
- O R-squared varia de 0.137 a 0.540, indicando que a temperatura explica de 13.7% a 54% da variabilidade nas diferenças de precipitação.
- O maior poder explicativo ocorre no modelo HadGEM2_ES sob o cenário rcp85 (R-squared = 0.540), sugerindo que, nesse cenário, a relação entre temperatura e precipitação é mais forte e consistente.
- Todos os p-valores são < 0.05, indicando que a relação é estatisticamente significativa em todos os modelos e cenários.

### Interpretação dos Parâmetros

- **Coef_Temp:** Cada aumento de 1 unidade na diferença de temperatura está associado a uma redução na diferença de precipitação que varia de aproximadamente 90 a 188 unidades, dependendo do modelo e cenário.
- **Intercept:** Valor esperado da diferença de precipitação quando a diferença de temperatura é zero.
- **R-squared:** Proporção da variabilidade da precipitação explicada pela temperatura.
- **p-value:** Todos indicam forte significância estatística da relação.


```python
from statsmodels.tsa.stattools import ccf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Número máximo de defasagens
max_lag = 5

# Loop por modelo e cenário
for (modelo, scenario), grupo in df_diff.groupby(['model', 'scenario']):
    dados = grupo.dropna(subset=['precipitation_diff', 'temperature_diff'])

    if len(dados) > max_lag:
        temp_diff = dados['temperature_diff'].values
        precip_diff = dados['precipitation_diff'].values

        # Calcular CCF apenas para lags positivos
        ccf_values = ccf(temp_diff, precip_diff)[:max_lag + 1]
        lags = np.arange(0, max_lag + 1)

        # Plot
        plt.figure(figsize=(6, 4))
        markerline, stemlines, baseline = plt.stem(lags, ccf_values)
        plt.setp(stemlines, 'color', 'blue')
        plt.setp(markerline, 'color', 'blue')
        plt.title(f'CCF - {modelo} | {scenario}')
        plt.xlabel('Lag (anos)')
        plt.ylabel('Correlação')
        plt.axhline(0, color='gray', linestyle='--')
        plt.ylim(-1, 1)
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Grupo {modelo} | {scenario} não possui dados suficientes para CCF.")

```

### Análise de Correlação Cruzada (CCF) - Interpretação dos Resultados

- Todos os modelos e cenários apresentam uma correlação negativa significativa no lag 0, consistente com os resultados anteriores de correlação e regressão.
- Observa-se uma inversão de sinal nos lags 1 e 2, com correlação positiva, sugerindo que após anos de redução de precipitação associada ao aumento de temperatura, há tendência de recuperação ou ajuste nos anos seguintes.
- Este padrão é consistente entre modelos e cenários, sendo mais evidente no HadGEM2_ES e ACCESS1_3 sob rcp85, onde a correlação em lag 1 atinge valores próximos de 0.4 a 0.5.
- A partir dos lags 3 a 5, as correlações diminuem e se aproximam de zero, indicando que o efeito defasado não persiste além de dois anos.
- Este comportamento sugere que os impactos das variações de temperatura sobre a precipitação não são puramente contemporâneos, mas podem ter efeitos de curto prazo (1 a 2 anos) nas dinâmicas do sistema climático representado pelos modelos.

### Resumo dos Padrões Observados

- **Lag 0:** Correlação negativa consistente — aumento de temperatura leva à redução imediata da precipitação.
- **Lag 1:** Correlação positiva moderada — possível efeito de recuperação ou compensação hídrica no ano seguinte.
- **Lag 2:** Correlação positiva, mas mais fraca que no lag 1.
- **Lags 3 a 5:** Correlação próxima de zero — desaparecimento do efeito defasado.

Este padrão é robusto em todos os modelos e cenários, mas mais acentuado nos cenários de maior forçamento radiativo (rcp85).
